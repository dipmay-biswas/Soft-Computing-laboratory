1. Let A(x) and B(x) be two fuzzy sets of speed limit on a highway, 
defined by the following membership functions:

```python
import matplotlib.pyplot as plt
import numpy as np
```

a. Fuzzy set A: “Low Speed Limit”\
μA(x) = {1, if x < 30 (low speed)\
= x − 30 / 50 − 30 , if 30 < x < 50 (moderately low speed)\
= x − 50 /70 − 50 , if 50 < x < 70 (higher low speed)\
= 0, if x > 70 (Nnot low speed)}

```python
def mu_A(x):
 if x < 30:
     return 1
 elif 30 <= x < 50:
     return (x - 30) / (50 - 30)
 elif 50 <= x < 70:
     return (x - 50) / (70 - 50)
 else:
     return 0
```

b. Fuzzy set B: “High Speed Limit”\
μB(x) = {0, if x < 60 (Not high speed)\
= x − 60 / 80 − 60, if 60 < x < 80 (moderately high speed)\
= x − 80 / 100 − 80 , if 80 < x < 100 (higher high speed)\
= 1, if x > 100 (high speed)}

```python
def mu_B(x):
 if x < 60:
     return 0
 elif 60 <= x < 80:
     return (x - 60) / (80 - 60)
 elif 80 <= x < 100:
     return (x - 80) / (100 - 80)
 else:
     return 1
```

Assume universe of discourse is the speed of cars {10 km/hr, 
20km/hr...........150 km/hr}\
Perform the following operations to obtain the resultant set and plot 
the same:\
i. Union – C(x) = (A(x) B(x)); draw the membership curve of fuzzy Set ∪
C(x)

```python
def mu_union(x):
 return max(mu_A(x), mu_B(x))
x = np.arange(10, 151, 10)
membership_1u = [mu_union(val) for val in x]
plt.plot(x, membership_1u, label=" A(x) ∪ B(x)")
plt.xlabel("Speed (km/hr)")
plt.ylabel("Membership Degree")
plt.title("Membership Curve for Fuzzy Set")
plt.legend()
plt.grid(True)
plt.show()
```

ii. Intersection – D(x) = (A(x) B(x)) draw the membership curve of ∩
fuzzy Set D(x)

```python
def mu_inter(x):
 return min(mu_A(x), mu_B(x))
membership_1i = [mu_inter(val) for val in x]
plt.plot(x, membership_1i, label="A(x) ∩ B(x)")
plt.xlabel("Speed (km/hr)")
plt.ylabel("Membership Degree")
plt.title("Membership Curve for Fuzzy Set")
plt.legend()
plt.grid(True)
plt.show()
```

2. Let C(x) and D(x) be two fuzzy sets of temperature (Celsius) in a city, defined by the following membership functions:

a. Fuzzy set C: “Cold”\
μC(x) ={0, if x < -10 (extremely cold)\
= (x−(−10) / (0−(−10) ,if -10 < x < 0 (very cold)\
= 1, if 0 < x < 5 (cold)\
= 0, if x > 5 (not cold)}


```python
def mu_C(x):
 if x < -10:
     return 0
 elif -10 <= x < 0:
     return (x - (-10)) / (0 - (-10))
 elif 0 <= x < 5:
     return 1
 else:
     return 0
```

b. Fuzzy set D: “Warm”\
μD(x) = {0, if x < 25 (Not warm)\
= x−25 / 35−25, if 25 < x < 35 (moderately warm)\
= 1, if 35 < x < 40 (warm)\
= 0, if x > 40 (extremely warm)}


```python
def mu_D(x):
 if x < 25:
     return 0
 elif 25 <= x < 35:
     return (x - 25) / (35 - 25)
 elif 35 <= x < 40:
     return 1
 else:
     return 0
```

Assume universe of discourse of Temperature U ={-100C, -50C, ....... 
500C}\
Perform the following operations to obtain the result:\
Max-Min Composition – F (x) = C(x) D(x)

```python
y = np.arange(-10,51,5)
C = [mu_C(val) for val in y]
D = [mu_D(val) for val in y]
```

```python
n = len(C)
mmc_i = []
for valC in C:
     temp = []
     for valD in D:
         temp.append(min(valD,valC))
     mmc_i.append(temp)
mmc_i
```

```python
mmcC = []
for i in range(n):
     tempC = []
     for j in range(n):
         tempC.append(min(mmc_i[i][j],C[j]))
     mmcC.append(max(tempC))
mmcC
```

```python
mmcD = []
for i in range(n):
     tempD = []
     for j in range(n):
         tempD.append(min(mmc_i[i][j],D[j]))
     mmcD.append(max(tempD))
mmcD
```

Draw the membership curve of F(x)

```python
plt.plot(y, mmcC, label="F(x) = C(x).D(x)")
plt.xlabel("Temperature (*C)")
plt.ylabel("Membership Degree")
plt.title("Membership Curve for Fuzzy Set")
plt.legend()
plt.grid(True)
plt.show()
```
