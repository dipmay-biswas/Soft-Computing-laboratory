```python
# Import necessary libraries
import numpy as np  # Import the NumPy library for numerical operations
from sklearn.model_selection import KFold  # Import KFold for cross-validation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report  # Import evaluation metrics
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
from keras.models import Sequential  # Import Sequential model from Keras
from keras.layers import Dense  # Import Dense layer from Keras for building neural networks
from keras.optimizers import SGD  # Import the Stochastic Gradient Descent (SGD) optimizer from Keras

# Define a function to create the neural network model
def create_model(hidden_layers, nodes_per_layer, learning_rate):
    # Create a sequential model
    model = Sequential()
    # Add the input layer with 'input_dim' input features and 'nodes_per_layer' neurons
    model.add(Dense(nodes_per_layer, input_dim=49, activation='sigmoid'))
    
    # Add hidden layers (the number of layers is determined by 'hidden_layers')
    for _ in range(hidden_layers - 1):
        model.add(Dense(nodes_per_layer, activation='sigmoid'))
    
    # Add the output layer with 1 neuron and sigmoid activation for binary classification
    model.add(Dense(1, activation='sigmoid'))
    # Compile the model with specified loss function, optimizer, and metrics
    optimizer = SGD(learning_rate=learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['accuracy'])
    
    return model

# Generate training data
num_train_examples = 500  # Number of training examples
X_train = np.random.randint(0, 2, size=(num_train_examples, 7, 7))  # Generate random 7x7 binary images
y_train = np.repeat(np.arange(10), num_train_examples // 10)  # Generate labels (0-9)

# Reshape input to be 2D (flattening the 7x7 images into 1D)
X_train = X_train.reshape(num_train_examples, 49)

# Model configuration
hidden_layers = 2  # Number of hidden layers in the neural network
nodes_per_layer = 50  # Number of neurons in each hidden layer
learning_rate = 0.1  # Learning rate for the optimizer
num_epochs = 100  # Number of training epochs

# Create the neural network model using the defined function
model = create_model(hidden_layers, nodes_per_layer, learning_rate)

# Import necessary libraries
import numpy as np  # Import the NumPy library for numerical operations
from sklearn.model_selection import KFold  # Import KFold for cross-validation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report  # Import evaluation metrics
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
from keras.models import Sequential  # Import Sequential model from Keras
from keras.layers import Dense  # Import Dense layer from Keras for building neural networks
from keras.optimizers import SGD  # Import the Stochastic Gradient Descent (SGD) optimizer from Keras

# Define a function to create the neural network model
def create_model(hidden_layers, nodes_per_layer, learning_rate):
    # Create a sequential model
    model = Sequential()
    # Add the input layer with 'input_dim' input features and 'nodes_per_layer' neurons
    model.add(Dense(nodes_per_layer, input_dim=49, activation='sigmoid'))
    
    # Add hidden layers (the number of layers is determined by 'hidden_layers')
    for _ in range(hidden_layers - 1):
        model.add(Dense(nodes_per_layer, activation='sigmoid'))
    
    # Add the output layer with 1 neuron and sigmoid activation for binary classification
    model.add(Dense(1, activation='sigmoid'))

    # Compile the model with specified loss function, optimizer, and metrics
    optimizer = SGD(learning_rate=learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['accuracy'])
    
    return model

# Generate training data
num_train_examples = 500  # Number of training examples
X_train = np.random.randint(0, 2, size=(num_train_examples, 7, 7))  # Generate random 7x7 binary images
y_train = np.repeat(np.arange(10), num_train_examples // 10)  # Generate labels (0-9)

# Reshape input to be 2D (flattening the 7x7 images into 1D)
X_train = X_train.reshape(num_train_examples, 49)

# Model configuration
hidden_layers = 2  # Number of hidden layers in the neural network
nodes_per_layer = 50  # Number of neurons in each hidden layer
learning_rate = 0.1  # Learning rate for the optimizer
num_epochs = 100  # Number of training epochs

# Create the neural network model using the defined function
model = create_model(hidden_layers, nodes_per_layer, learning_rate)

# Train the model on the training data for 'num_epochs' epochs (verbose=0 means no verbose output)
history = model.fit(X_train, y_train, epochs=num_epochs, verbose=0)

# Plot loss vs. iterations to study convergence
plt.plot(history.history['loss'])  # Plot the loss values from training
plt.xlabel('Iteration')  # Label for the x-axis
plt.ylabel('Mean Squared Error Loss')  # Label for the y-axis
plt.title('Loss vs. Iterations')  # Title of the plot
plt.show()  # Display the plot

# N-fold cross-validation
kf = KFold(n_splits=5)  # Create a 5-fold cross-validation object
scores = []  # Initialize a list to store accuracy scores for each fold

# Loop through the folds (train and evaluate models for each fold)
for train_index, test_index in kf.split(X_train):
    # Split data into training and testing sets for cross-validation
    X_train_cv, X_test_cv = X_train[train_index], X_train[test_index]
    y_train_cv, y_test_cv = y_train[train_index], y_train[test_index]

    # Create a new model for each fold with the same architecture
    model = create_model(hidden_layers, nodes_per_layer, learning_rate)

    # Train the model on the training data for this fold
    model.fit(X_train_cv, y_train_cv, epochs=num_epochs, verbose=0)

    # Predict on the test data for this fold
    y_pred = model.predict(X_test_cv)
    y_pred_binary = np.round(y_pred).flatten()  # Convert predicted probabilities to binary values

    # Calculate accuracy for this fold
    accuracy = accuracy_score(y_test_cv, y_pred_binary)

    # Calculate confusion matrix and classification report for this fold
    confusion = confusion_matrix(y_test_cv, y_pred_binary)
    report = classification_report(y_test_cv, y_pred_binary, target_names=[str(i) for i in range(10)])

    # Append accuracy to the list of scores
    scores.append(accuracy)

    # Print accuracy, confusion matrix, and classification report for this fold
    print("Accuracy for Fold:")
    print(accuracy)
    print("Confusion Matrix for Fold:")
    print(confusion)
    print("Classification Report for Fold:")
    print(report)
    print("-" * 50)
# Calculate and print the mean and standard deviation of accuracy across all folds
print("K-fold CV Accuracy Mean:", np.mean(scores))
print("K-fold CV Accuracy Standard Deviation:", np.std(scores))
```
